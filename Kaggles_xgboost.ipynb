```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.preprocessing import MinMaxScaler
```

```python
# Load the dataset
train_data = pd.read_csv("/content/Trout_training.csv")
test_data = pd.read_csv("/content/Trout_testing_features.csv")
train_data
```

```python
# Pivot the training and test data
train_data_p = train_data.pivot(index='Date_Time', columns='Depth_m', values='Water_Temp_C').reset_index()
test_data_p = test_data.pivot(index='Date_Time', columns='Depth_m', values='Water_Temp_C').reset_index()
```

```python
train_data_p
```

```python
train_data_p['Date_time'] = pd.to_datetime(train_data_p['Date_Time'])
test_data_p['Date_time'] = pd.to_datetime(test_data_p['Date_Time'])
```

```python
# Extract hour, day, month, and year from date_time
train_data_p['hour'] = train_data_p['Date_time'].dt.hour
train_data_p['day'] = train_data_p['Date_time'].dt.day
train_data_p['month'] = train_data_p['Date_time'].dt.month
train_data_p['year'] = train_data_p['Date_time'].dt.year

test_data_p['hour'] = test_data_p['Date_time'].dt.hour
test_data_p['day'] = test_data_p['Date_time'].dt.day
test_data_p['month'] = test_data_p['Date_time'].dt.month
test_data_p['year'] = test_data_p['Date_time'].dt.year
```
